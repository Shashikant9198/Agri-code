```python

"""1.IMPORTING OS (OPERATING SYSTEM) TO READ IMAGES IN FOLDERS"""

import os

"""2.WE NEED TO STORE LIST OF DIRECTORY IN SOME LIST , SO Leaf_Smut NAMED LIST DECLARED AND 
USING os.listdir METHOD WE CAN READ DIRECTORY OF ALL IMAGES IN AND STORE IN THE LSIT """

Leaf_Smut=os.listdir("C:/Users/Dell/Desktop/Leaf smut")
print(Leaf_Smut)

"""3.BASICALLY OUR MACHINE KNOWS ONL NUMERIC VALUES AND MACHINE DONT WORK ON IMAGE DIRECTLY, 
SO, Imageio is a Python library that provides an easy interface to read and write a wide range of IMAGE DATA... """

import imageio
import numpy as np

#CREATING EMPTY LIST Image=[] TO STORE READ IMAGES 
Image=[]

#ITERATING Leaf_Smut FOR 0 TO 39 IMAGES OF LEAFSMUT DISEASE AND 
#STORING IN EMPTY LIST Image=[]
#BY USING append FUNCTION.
for i in Leaf_Smut:
    img = imageio.imread( 'C:/Users/Dell/Desktop/Leaf smut/'+i)
    Image.append(img)
    #Image=np.asarray(img)
    print(Image)
Image   




"""4.OpenCV (Open Source Computer Vision Library) is an open source computer vision and machine learning software library"""
import cv2

"""5.NOW WE HAVE TO RESIZE ALL IMAGES FROM OUR FOLDER , SO..WE ITERATED USING FOR LOOP AND APPENDED IN THE EMPTY LIST Image2=[]"""
Image2=[]
for j in range(40):
    Image[j].width=Image[j].shape[1]
    
    #OUR ALL IMAGES ARE OF DIFFERENT RESOLUTIONS , SO THIS GIVES US ARRAYS MATRIX OF VARIABLE DIMENTIONS 
    #AND THE FEATURE ARRAY VECTOR MUST BE OF SAME DIMENTIONS FOR EACH IMAGE ,SO WE RESIZED EVERY IMAGE TO 480*1000 SIZE
    Image[j].height =480
    Image[j].width = 1000
    dim = (Image[j].width, Image[j].height)
    
    #cv2.resize FUNCTION IS USED RESIZE IMAGES
    resized = cv2.resize(Image[j], dim, interpolation = cv2.INTER_AREA)
    Image2.append(resized)
    #Image2=array(Image2)
    print(j)
print('Resized Dimensions : ',resized.shape)
print('Original Dimensions : ',Image[j].shape)

"""6.WE PREVIOUSLY USED OPENCV AS CV2 AND ITS DRAWBACK IS, CV2 CONVERTS OUR COLOUR IMAGE TO BLUE GREEN RED FORMAT 
AND WE WANT THE IDEAL RED GREEN BLUE PIXEL VALUE SEQUENCE IN ARRAY FOR EX. [B G R] TO [R G B] WHERE R G B REPRESENTS 
PIXEL VALUE OF COLURS """
#USING OPENCV cv2.cvtColor
Image2[j] = cv2.cvtColor(Image2[j], cv2.COLOR_BGR2RGB)
cv2.imshow("Resized image", Image2[j])
cv2.waitKey(0)
cv2.destroyAllWindows()

#CONVERTING LIST Image2 TO NUMPY ARRAY FOR MANIPULATING EASILY
Image2=np.array(Image2)







######################################################





"""1.IMPORTING OS (OPERATING SYSTEM) TO READ IMAGES IN FOLDERS"""

import os

"""2.WE NEED TO STORE LIST OF DIRECTORY IN SOME LIST , SO Brown_spot  LIST DECLARED AND 
USING os.listdir METHOD WE CAN READ DIRECTORY OF ALL IMAGES IN AND STORE IN THE LSIT """
 
Brown_spot=os.listdir("C:/Users/Dell/Desktop/Brown spot")
print(Brown_spot)

"""3.BASICALLY OUR MACHINE KNOWS ONL NUMERIC VALUES AND OUR MACHINE DONT WORK ON IMAGE DIRECTLY, 
SO, Imageio is a Python library that provides an easy interface to read and write a wide range of IMAGE DATA... """

import imageio
import numpy as np
#CREATING EMPTY LIST Image3=[] TO STORE READ IMAGES
Image3=[]
#ITERATING Brown_spot FOR 40 TO 79 IMAGES OF LEAFSMUT DISEASE AND 
#STORING IN EMPTY LIST Image3=[]
#BY USING append FUNCTION.

for i in Brown_spot:
    img1 = imageio.imread( 'C:/Users/Dell/Desktop/Brown spot/'+i)
    Image3.append(img1)
    #Image=np.asarray(img)
    print(Image3)
Image3 




"""4.OpenCV (Open Source Computer Vision Library) is an open source computer vision and machine learning software library"""
import cv2
"""5.NOW WE HAVE TO RESIZE ALL IMAGES FROM OUR FOLDER ,SO..WE ITERATED USING FOR LOOP AND APPENDED IN THE EMPTY LIST Image4=[]"""
Image4=[]
for j in range(40):
    Image3[j].width=Image3[j].shape[1]
    #OUR ALL IMAGES ARE OF DIFFERENT RESOLUTIONS , SO THIS GIVES US ARRAYS MATRIX OF VARIABLE DIMENTIONS 
    #AND THE FEATURE ARRAY VECTOR MUST BE OF SAME DIMENTIONS FOR EACH IMAGE ,SO WE RESIZED EVERY IMAGE TO 480*1000 SIZE
    Image3[j].height =480
    Image3[j].width = 1000
    dim1 = (Image3[j].width, Image3[j].height)
    
    #cv2.resize FUNCTION IS USED RESIZE IMAGES
    resized1 = cv2.resize(Image3[j], dim1, interpolation = cv2.INTER_AREA)
    Image4.append(resized1)
    #Image2=array(Image2)
    print(j)
print('Resized Dimensions : ',resized1.shape)
print('Original Dimensions : ',Image3[j].shape)

"""6.WE PREVIOUSLY USED OPENCV AS CV2 AND ITS DRAWBACK IS, CV2 CONVERTS OUR COLOUR IMAGE TO BLUE GREEN RED FORMAT 
AND WE WANT THE IDEAL RED GREEN BLUE PIXEL VALUE SEQUENCE IN ARRAY FOR EX. [B G R] TO [R G B] WHERE R G B REPRESENTS 
PIXEL VALUE OF COLURS """
Image4[j] = cv2.cvtColor(Image4[j], cv2.COLOR_BGR2RGB)
cv2.imshow("Resized image", Image4[j])
cv2.waitKey(0)
cv2.destroyAllWindows()
Image4=np.array(Image4)



#############################










"""1.IMPORTING OS (OPERATING SYSTEM) TO READ IMAGES IN FOLDERS"""
import os
#for i in range 
"""2.WE NEED TO STORE LIST OF DIRECTORY IN SOME LIST , SO Bacterial_leaf_blight LIST DECLARED AND 
USING os.listdir METHOD WE CAN READ DIRECTORY OF ALL IMAGES IN AND STORE IN THE LSIT """
Bacterial_leaf_blight=os.listdir("C:/Users/Dell/Desktop/Bacterial leaf blight")
print(Bacterial_leaf_blight)

"""3.BASICALLY OUR MACHINE KNOWS ONL NUMERIC VALUES AND OUR MACHINE DONT WORK ON IMAGE DIRECTLY, 
SO, Imageio is a Python library that provides an easy interface to read and write a wide range of IMAGE DATA... """
import imageio
import numpy as np
#CREATING EMPTY LIST Image3=[] TO STORE READ IMAGES
Image5=[]

#ITERATING Bacterial_leaf_blight FOR 80 TO 119 IMAGES OF LEAFSMUT DISEASE AND 
#STORING IN EMPTY LIST Image5=[]
#BY USING append FUNCTION.

for i in Bacterial_leaf_blight:
    img2 = imageio.imread( 'C:/Users/Dell/Desktop/Bacterial leaf blight/'+i)
    Image5.append(img2)
    #Image=np.asarray(img)
    print(Image5)
Image5   


"""4.OpenCV (Open Source Computer Vision Library) is an open source computer vision and machine learning software library"""
import cv2
"""5.NOW WE HAVE TO RESIZE ALL IMAGES FROM OUR FOLDER ,SO..WE ITERATED USING FOR LOOP AND APPENDED IN THE EMPTY LIST Image6=[]"""
Image6=[]
for j in range(40):
    Image5[j].width=Image5[j].shape[1]
    #OUR ALL IMAGES ARE OF DIFFERENT RESOLUTIONS , SO THIS GIVES US ARRAYS MATRIX OF VARIABLE DIMENTIONS 
    #AND THE FEATURE ARRAY VECTOR MUST BE OF SAME DIMENTIONS FOR EACH IMAGE ,SO WE RESIZED EVERY IMAGE TO 480*1000 SIZE
    Image5[j].height =480
    Image5[j].width = 1000
    dim1 = (Image5[j].width, Image5[j].height)
    
    #cv2.resize FUNCTION IS USED RESIZE IMAGES
    resized1 = cv2.resize(Image5[j], dim1, interpolation = cv2.INTER_AREA)
    Image6.append(resized1)
    #Image2=array(Image2)
    print(j)
print('Resized Dimensions : ',resized1.shape)
print('Original Dimensions : ',Image5[j].shape)

"""6.WE PREVIOUSLY USED OPENCV AS CV2 AND ITS DRAWBACK IS, CV2 CONVERTS OUR COLOUR IMAGE TO BLUE GREEN RED FORMAT 
AND WE WANT THE IDEAL RED GREEN BLUE PIXEL VALUE SEQUENCE IN ARRAY FOR EX. [B G R] TO [R G B] WHERE R G B REPRESENTS 
PIXEL VALUE OF COLURS """

Image5[j] = cv2.cvtColor(Image6[j], cv2.COLOR_BGR2RGB)
cv2.imshow("Resized image", Image5[j])
cv2.waitKey(0)
cv2.destroyAllWindows()
Image6=np.array(Image6)


####################################################

"""LABEL ENCODING FOR THREE ARRAYS AS 0,1,2"""
from sklearn import preprocessing 
labels=['Image2','Image4','Image6']
encoder=preprocessing.LabelEncoder()
encoder.fit(labels)
for a,item in enumerate(encoder.classes_):
    print(item,'=>',a)




####################################################



"""TO CREATE A FEATURE VECTOR TO PASS THROUGH MODEL , MERGING OF THREE ARRAY"""
Imgfnl=np.vstack([Image2,Image4,Image6])

"""CONSIDERING Imgfnl AS X"""
X=Imgfnl
X
"""CREATING Y FUNCTION OF LABELS 0,1,2"""
y = []
for a in range(120):
    if a<40:
        y.append(0)
    elif 40<=a<80:
        y.append(1)
    else:
        y.append(2)
        
#CONVERTING LIST TO NUMPY ARRAY         
Y=np.array(y)      

#######################################################

#DATA VALIDATION INTO TEST DATA, TRAIN DATA WHERE SHUFFLE IS TRUE MEANS RANDOM DATA IS TAKEN FROM DATASET
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(X, Y, test_size = 0.1, random_state = 0,shuffle="true")
xTrain.shape#PRINTS THE SHAPE OF xTrain

######################################################
#TENSORFLOW
"""TensorFlow is an open source library for numerical computation and large-scale machine learning. 
TensorFlow bundles together a slew of machine learning and deep learning (aka neural networking) models and 
algorithms and makes them useful..."""
import tensorflow as tf


#KERAS
"""Keras is a high-level neural networks API, written in Python and capable of running on top of TensorFlow"""



#SEQUENSIAL
"""Sequential ALLOWS USER TO DEFINE LAYERS OF CONVOLUTIONAL NEURAL NETWORKS """
from tensorflow.keras.models import Sequential




#MAXPOOLING2D LAYER
"""MaxPooling2D IS USED TO DOWNSAMPLE THE FEATURE MATRIX TO REDUCED DIMENTIONS , IT EXTRACTS THE HIGHEST FEATURES FROM MARTIX 
AND CREATE ONE MATRIX CONTAINING HIGHLIGHTED FEATURE VALUES OF LESS DIMENTION  """
#CONV2D
"""THIS IS SIMPLE CONVOLUTIONAL LAYER WHICH CONNECTS WEIGHTS INPUT TO NEURONS AND EACH CONV LAYER NEURON TO NEAR BY 
NEURONS OF NEXT LAYER"""
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten,Conv2D,MaxPooling2D



#ADAM OPTIMISER
"""ADAM OPTIMISER IS USED HERE FOR GRADIENT DESCENT OVER FUNCTION AND LEARNING RATE OF 0.001"""
from tensorflow.keras.optimizers import Adam


  
    
"""  ARCHITECTURE OF CNN MODEL  """    
model =Sequential()
X=X

#LAYER_1
model.add(Conv2D(10,(5,5),input_shape=X.shape[1:])     )
model.add(Activation("relu"))
#RECTIFIED LINEAR UNIT
"""ReLU is linear (identity) for all positive values, and zero for all negative values, AND ITS A ACTIVATION FUNCTION """
model.add(MaxPooling2D(pool_size=(2,2)))


#LAYER_2
model.add(Conv2D(6,(5,5),input_shape=X.shape[1:])     )
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))


#LAYER_3
model.add(Conv2D(4,(5,5),input_shape=X.shape[1:])     )
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))


#LAYER_4
#model.add(Conv2D(3,(5,5),input_shape=X.shape[1:])     )
#model.add(Activation("relu"))
#model.add(MaxPooling2D(pool_size=(2,2)))


model.add(Flatten())                     #FLATEEN LAYER
"""The role of the Flatten layer in Keras is super simple: 
A flatten operation on a tensor reshapes the tensor to have the shape that is equal to 
the number of elements contained in tensor non including the batch dimension """


model.add(Dense(3))                      #DENSE LAYER
"""DENSE LAYER IS FULLY CONNECTED TO OUTPUT LAYER """

#LAYER_LAST
model.add(Dense(1))
model.add(Activation("sigmoid"))     #SIGMOID ACTIVATION FUNCTION IS FOR FULLY CONNECTED LAYER TO OUTPUT



model.compile(Adam(lr=.001),loss="binary_crossentropy",metrics=['accuracy'])#ACCURACY METROCS GIVES ACCURACY,LOSS,VAL_ACC,VAL_LOSS


# GIVING MODEL TO TRAIN # 
model.fit(X,Y,batch_size=2,epochs=5,validation_split=0.2,shuffle=True) 

# HERE TAKING SUMMARY OF OUR NETWORK#
model.summary 
```
